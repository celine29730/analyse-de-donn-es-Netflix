# Analyse statistique de la table film

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('film1.csv')
dataset.head()
```

le dataset que l'on étudie contient l'identifiant du film, le titre, la date d'ajout sur Netflix,l'année de sortie, 
la note attribuée au film, la durée du film et la description.

```python
dataset.shape
```

**Etude de la variable quantitative release_year**

```python
dataset.release_year.describe().round(1)
```

**les films du dataset sont sortie en moyenne durant l'année 2012, le film le plus ancien est de 1942, 25% des films 
sont sortis avant 2011 et 50% des films sont sortis avant 2016. 
Le film le plus récent date de 2019.**

```python
dataset.release_year.plot(kind = "hist")
```

```python
dataset.boxplot(column = "release_year")
```

**Etude de la variable catégorielle rating**


```python
dataset.rating.describe()
```

```python
dataset.rating.value_counts()

pd.crosstab(dataset.rating, "freq")  
```

```python
cam = pd.crosstab(dataset.rating, "freq")
cam.plot.pie(subplots=True, figsize = (8,8))
```

**présentation des statistiques de base (la moyenne, l'écart-type, médiane, min et max) de la variable Année de 
sortie des films en fonction des notes attribuées**

```python
dataset.groupby("rating")["release_year"].agg([np.mean, np.std, np.median, np.min, np.max])
```
